# Imports

```python
# CÉLULA 1: Importando bibliotecas básicas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
from PIL import Image

print("Pronto! Vamos comparar os tipos de dados.")
```

# Dados

```python
# CÉLULA 2: O Dado Tabular (O que estamos acostumados)
print("--- 1. DADOS TABULARES (Tabelas) ---")
print("Cenário: Dados de vendas de uma loja.")

# Criando uma tabela simples
dados_loja = {
    'Cliente': ['Ana', 'Bruno', 'Carlos'],
    'Idade': [28, 34, 45],
    'Valor_Gasto_R$': [150.50, 320.00, 89.90]
}
tabela = pd.DataFrame(dados_loja)

display(tabela)
print("\nConclusão: O dado tabular é estruturado, fácil para o cérebro humano ler em linhas e colunas.")
```

```python
# CÉLULA 3: O Dado de Imagem (O mundo da Visão Computacional)
print("\n--- 2. DADOS DE IMAGEM (Matrizes) ---")
print("Cenário: Como o computador 'lê' uma foto?")

# Baixando e mostrando uma imagem simples de uma maçã
url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/apple.jpg"
urllib.request.urlretrieve(url, "apple.jpg")

# Abrindo a imagem e reduzindo bem o tamanho para a matriz não ficar gigante na tela
imagem = Image.open("apple.jpg").resize((5, 5))

# Mostrando a imagem visualmente
plt.figure(figsize=(2,2))
plt.imshow(imagem)
plt.axis('off')
plt.title("O que nós vemos")
plt.show()
```

```python
# Transformando a imagem na matriz matemática que o computador enxerga
matriz_imagem = np.array(imagem)
print("\nO que o COMPUTADOR vê (Matriz de Pixels):")
print(f"Formato: {matriz_imagem.shape} -> (Altura, Largura, Canais de Cor RGB)")
print("\nValores numéricos de um pedaço da imagem (intensidade das cores de 0 a 255):")
print(matriz_imagem[0:2, 0:2]) # Mostrando apenas um pedacinho 2x2
```

```python
# CÉLULA 4: PROGRAMAÇÃO TRADICIONAL (Regras Fixas)
print("--- PROGRAMAÇÃO TRADICIONAL ---")
print("Cenário: Dar desconto de 10% para compras acima de R$ 250,00.")

# Na programação tradicional, o PROGRAMADOR escreve a regra (IF / ELSE)
def sistema_de_desconto_tradicional(valor_da_compra):
    # Olha a regra escrita em código aqui:
    if valor_da_compra >= 250.00:
        return "✅ Desconto Aprovado!"
    else:
        return "❌ Sem Desconto."

# Testando
compra_cliente = 250.00
resultado = sistema_de_desconto_tradicional(compra_cliente)
print(f"Cliente gastou R$ {compra_cliente} -> O sistema disse: {resultado}")
print("\nProblema: Se a regra do negócio mudar amanhã, o programador tem que abrir o código e reescrever!")
```

```python
# CÉLULA 5: MACHINE LEARNING (A máquina descobre a regra)
from sklearn.tree import DecisionTreeClassifier

print("\n--- MACHINE LEARNING ---")
print("Cenário: A máquina descobre a regra sozinha olhando o histórico de vendas.")

# DADOS (Histórico de valores de compras)
historico_compras = [[50], [120], [180], [210], [300], [500]]

# RESULTADOS (O que aconteceu no passado: 0 = Sem Desconto, 1 = Teve Desconto)
# Note que quem gastou 210, 300 e 500 ganhou desconto, mas a máquina não sabe o motivo ainda.
historico_descontos = [0, 0, 0, 1, 1, 1]

# Nós NÃO escrevemos IF ou ELSE. Nós damos os dados para o Algoritmo:
modelo_ia = DecisionTreeClassifier()
modelo_ia.fit(historico_compras, historico_descontos) # O comando 'fit' é o treinamento (o aprendizado)

print("IA treinada com os dados do passado! Ela já descobriu a regra matemática sozinha.")

# Testando com novos clientes
novo_cliente_1 = [[190]]
novo_cliente_2 = [[220]]

if modelo_ia.predict(novo_cliente_1)[0] == 1:
    print(f"Novo cliente gastou R$ 190 -> IA disse: ✅ Desconto Aprovado!")
else:
    print(f"Novo cliente gastou R$ 190 -> IA disse: ❌ Sem Desconto.")

if modelo_ia.predict(novo_cliente_2)[0] == 1:
    print(f"Novo cliente gastou R$ 220 -> IA disse: ✅ Desconto Aprovado!")
else:
    print(f"Novo cliente gastou R$ 220 -> IA disse: ❌ Sem Desconto.")
```
